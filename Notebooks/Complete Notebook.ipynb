# Outlier Detection

```python
X = healthtrain.drop('target', axis=1)
y = healthtrain['target']
```

#import the implementation of this algorihm from sklearn
from sklearn.ensemble import IsolationForest

#Use the algorithm for outlier detection, then use it to predict each point
#Any point labelled as -1 is an outlier
isolation_forest = IsolationForest(max_samples=150, random_state = 1, contamination= 0.07)
preds = isolation_forest.fit_predict(X)
print(preds)
totalOutliers=0
for pred in preds:
    if pred == -1:
        totalOutliers=totalOutliers+1
print("Total number of outliers identified is: ",totalOutliers)

outlier_scores = isolation_forest.decision_function(X)
print(outlier_scores)

outlier_threshold = np.percentile(outlier_scores, 100 * 0.07)
print(outlier_threshold)

healthtrain = healthtrain[outlier_scores >= outlier_threshold]
healthtrain.info()

# Data Modelling

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import Normalizer
```

```python
X = healthtrain_cleaned.drop('target', axis=1)
y = healthtrain_cleaned['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from imblearn.over_sampling import ADASYN

adasyn = ADASYN()
X_resampled, y_resampled = adasyn.fit_resample(X_train, y_train)

normalizer = Normalizer(norm='l2')
X_resampled = normalizer.fit_transform(X_resampled)
X_test = normalizer.transform(X_test)
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X)
plt.title("Before Normalization")

plt.subplot(1, 2, 2)
sns.boxplot(data=X_resampled)
plt.title("After Normalization")
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1, 2, 1)
sns.set_style("darkgrid")
sns.histplot(healthtrain['target'])
plt.title("Before Upsampling")
plt.subplot(1, 2, 2)
sns.histplot(y_resampled)
plt.title("After Upsampling")
```

```python
y_resampled.shape
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

dt_clf = DecisionTreeClassifier()
parameters = {"criterion" : ['gini','entropy','log_loss']}
clf = GridSearchCV(dt_clf,param_grid=parameters)

clf.fit(X_resampled, y_resampled)
y_pred = clf.predict(X_test)
```

```python
print('Best parameter for decision tree:', clf.best_params_)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import recall_score, precision_score
#print(f1_score(y_test,y_pred))
print('Precision Score: ',precision_score(y_test,y_pred, average='weighted'))
print('Recall Score: ',recall_score(y_test,y_pred, average='weighted'))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn import tree

plt.figure(figsize=(50,30))
tree_plot = tree.plot_tree(clf.best_estimator_, class_names=['Low risk', 'Moderate risk', 'High risk'], filled = True, rounded=True)
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(clf, X_test, y_test)
```
