```python
def print_histogram(data, limit):
    # Filter values under the specified limit
    filtered_data = [value for value in data if value < limit]

    # Plotting the histogram
    plt.boxplot(filtered_data)  # Adjust the number of bins as needed
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram of Values under {}'.format(limit))

    # Display the histogram
    plt.show()

# Example usage
data = healthdata['x5']
limit = 0.1
print_histogram(data, limit)
```

```python
print(healthdata['x3'].value_counts())
print(healthdata['x14'].value_counts())
print(healthdata['target'].value_counts())
```

```python
sex_to_numeric = [0,1] #Male = 0 and Female = 1
blood_to_numeric = [0,1,2,3,4,5,6,7] #{O+:0, A+:1, B+:2, A-:3, B-:4, O-:5, AB+:6, AB-:7}
target_to_numeric = [1,2,3] #{Low risk:1, Moderate risk:2, High risk:3}

healthdata['x3'].replace(to_replace=['M','F'], value=sex_to_numeric, inplace=True)
healthdata['x14'].replace(to_replace=['O+','A+','B+','A-','B-','O-','AB+','AB-'], value=blood_to_numeric, inplace=True)
healthdata['target'].replace(to_replace=['Low risk', 'Moderate risk', 'High risk'], value=target_to_numeric, inplace=True)
#temp_df['target'].replace(to_replace=['Low risk', 'Moderate risk', 'High risk'], value=[1, 2, 3], inplace=True)
```
